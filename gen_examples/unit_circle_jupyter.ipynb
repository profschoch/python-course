```python
import numpy as np
import matplotlib.pyplot as plt
```

As usual, I've imported the libraries I need.  Now, I'll concentrate on creating the simple code to plot
the circle and the angled line.

```python
angle=np.pi*(1/4)

userx=np.append([0],[np.cos(angle)])
usery=np.append([0],[np.sin(angle)])
```

This is the angle for the line you want to plot.  Change the fraction in the parentheses, to get a different line.

```python
x=np.cos(np.arange(0.,2.2*np.pi,0.1))
y=np.sin(np.arange(0.,2.2*np.pi,0.1))
```

```python
plt.figure(figsize=(10,10))
plt.grid(True)
plt.plot(x,y,userx,usery)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Unit circle plotted with user defined angle")
```
